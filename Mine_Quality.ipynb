## Preprocessing

* ### predict silica percentage in iron

```python
# split x and y values

x = df.drop(df.columns[-1], axis=1).values
x
```

```python
y = df[df.columns[-1]].values
y = y.reshape(-1,1)
y
```

```python
x.shape,y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
x = sc.fit_transform(x)
y = sc.fit_transform(y)
```

### Training And Testing The Data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

### Evaluating the model

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred1 = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

mae1 = mean_absolute_error(y_test, y_pred1)
mse1 = mean_squared_error(y_test, y_pred1)
r2_score1 = r2_score(y_test, y_pred1)

print('Mean Absolute Error:', mae1)
print('Mean Squared Error:', mse1)
print('R^2 Score:', r2_score1)
```

#### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred2 = dt.predict(x_test)
```

```python
mae2 = mean_absolute_error(y_test, y_pred2)
mse2 = mean_squared_error(y_test, y_pred2)
r2_score2 = r2_score(y_test, y_pred2)

print('Mean Absolute Error:', mae2)
print('Mean Squared Error:', mse2)
print('R^2 Score:', r2_score2)
```

#### XGBOOST Regressor

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(x_train, y_train)
y_pred3 = xgb.predict(x_test)
```

```python
mae3 = mean_absolute_error(y_test, y_pred3)
mse3 = mean_squared_error(y_test, y_pred3)
r2_score3 = r2_score(y_test, y_pred3)

print('Mean Absolute Error:', mae3)
print('Mean Squared Error:', mse3)
print('R^2 Score:', r2_score3)
```

#### ANN

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
ann = tf.keras.models.Sequential()
```

### Input layers

```python
ann.add(Dense(250, input_dim = 22, kernel_initializer='normal',activation='relu'))
```

### Hidden layers

```python
ann.add(Dense(2000,activation = 'relu'))
ann.add(Dropout(0.1))  # avoid overfitting , 0.1 means that 10% of the input units will be randomly set to 0 during training. 
```

```python
ann.add(Dense(1000,activation = 'relu'))
ann.add(Dropout(0.1))  
```

```python
ann.add(Dense(500,activation = 'relu'))
ann.add(Dropout(0.1))  
```

```python
ann.add(Dense(250,activation = 'relu'))
ann.add(Dropout(0.1))  
```

### Output layer

```python
ann.add(Dense(1,activation = 'linear')) # for binary use sigmoid 
```

#### Making predictions and evaluating the model

```python
ann.compile(optimizer = 'adam',loss = 'mse',metrics = ['accuracy'])
ann.summary()
```

### Training ANN on training set

```python
ann.fit(x_train,y_train,batch_size=32,epochs=5,validation_split = 0.2)
```

```python
y_pred4 = ann.predict(x_test)
y_pred4 = (y_pred4 >= 0.5).astype(int)
y_test4 = (y_test >= 0.5).astype(int)

print(np.concatenate((y_pred4.reshape(-1, 1), y_test4.reshape(-1, 1)), axis=1))

```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

cm = confusion_matrix(y_test4,y_pred4)
cm
accuracy_score(y_test4,y_pred4)
```
